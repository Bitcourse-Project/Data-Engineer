```python
import pandas as pd

# 데이터 형식 : 테스트용 csv (구글 코랩 업로드해서 접근한 방식)
# 데이터 특성(열) : datetime(시간), open(시가), high(고가), low(저가), close(종가), vloume(거래량), NumTrades(거래 수)

# 파라미터 = period(고정), df(데이터프레임)
period = 14

# ------------------------------------------

# 여기서부터 로직에 들어가야 하는 부분

def RSI(df : pandas.DataFrame, period : int = 14):
    delta = df["close"].diff() # 종가열만 복사해서 diff 메서드 사용
    ups, downs = delta.copy(), delta.copy()
    ups[ups < 0] = 0
    downs[downs > 0] = 0

    AU = ups.ewm(com = period-1, min_periods = period).mean()
    AD = downs.abs().ewm(com = period-1, min_periods = period).mean()
    RS = AU / AD
    RSI = pandas.Series(100 - (100/(1+RS)))

      # Dataframe으로 써먹기 위해 Dataframe으로 바꿔준다.
    RSI = RSI.to_frame()

      # 해당 열의 이름을 RSI로 바꿔준다.
    RSI = RSI.rename(columns = {"close" : 'RSI'})

      # return pandas.Series(100 - (100/(1 + RS)), name = "RSI")
    return RSI

# while True:
#   data = pyupbit.get_ohlcv(ticker = "KRW-BTC", interval = "day") # 일봉 RSI 계산
#   now_rsi = rsi(data,14).iloc[-1]
#   print("현재 시각 : ",datetime.datetime.now(), "/ 현재 RSI : ",now_rsi)
#   time.sleep(60)
```
